#Задача 1
Написать функцию на вход которой подается строка, состоящая из латинских букв.
Функция должна вернуть количество гласных букв (a, e, i, o, u) в этой строке. Написать тесты для кода

```python
def count_vowels(s: str) -> int:
    vowels = {'a', 'e', 'i', 'o', 'u'}
    return sum(1 for char in s.lower() if char in vowels)


# Тесты
def test_count_vowels():
    assert count_vowels("hello") == 2
    assert count_vowels("HELLO") == 2
    assert count_vowels("xyz") == 0
    assert count_vowels("") == 0


if __name__ == "__main__":
    test_count_vowels()
```

# Задача 2
Написать функцию на вход, которой подается строка. Функция должна вернуть true, если
каждый символ в строке встречается только 1 раз, иначе должна вернуть false. Написать тесты для кода

```python
def all_unique_chars(s: str) -> bool:
    return len(s) == len(set(s))


# Тесты
def test_all_unique_chars():
    assert all_unique_chars("abcde") is True
    assert all_unique_chars("aabc") is False
    assert all_unique_chars("") is True
    assert all_unique_chars("AaBb") is True
    assert all_unique_chars("Hello") is False
    assert all_unique_chars("Python") is True


if __name__ == "__main__":
    test_all_unique_chars()
```

# Задача 3
Написать функцию, которая принимает положительное число и возвращает количество
бит равных 1 в этом числе. Написать тесты для кода

```python
def count_bits(n: int) -> int:
    if n < 0:
        raise ValueError("Число должно быть положительным")
    return bin(n).count("1")


# Тесты
def test_count_bits():
    assert count_bits(0) == 0
    assert count_bits(1) == 1
    assert count_bits(5) == 2  
    assert count_bits(7) == 3
    assert count_bits(8) == 1
    assert count_bits(255) == 8
    assert count_bits(1023) == 10


if __name__ == "__main__":
    test_count_bits()
```

# Задача 4
Написать функцию, которая принимает положительное число. Функция должна вернуть то,
сколько раз необходимо перемножать цифры числа или результат перемножения, чтобы
получилось число состоящее из одной цифры.
Например, для входного числа:
· 39 функция должна вернуть 3, так как 3*9=27 => 2*7=14 => 1*4=4
· 4 функция должна вернуть 0, так как число уже состоит из одной цифры
· 999 функция должна вернуть 4, так как 9*9*9=729 => 7*2*9=126 => 1*2*6=12 =>
1*2=2. Написать тесты для кода

```python
def magic(n: int) -> int:
    if n < 0:
        raise ValueError("Число должно быть положительным")

    steps = 0
    while n >= 10:
        product = 1
        for digit in str(n):
            product *= int(digit)
        n = product
        steps += 1

    return steps


def test_magic():
    assert magic(39) == 3
    assert magic(4) == 0
    assert magic(999) == 4
    assert magic(25) == 2
    assert magic(77) == 4
    assert magic(10) == 1
    assert magic(1) == 0


if __name__ == "__main__":
    test_magic()
```

# Задача 5
Написать функцию, которая принимает два целочисленных вектора одинаковой длины и
возвращает среднеквадратическое отклонение двух векторов. Написать тесты для кода

```python
import math

def mse(vec1: list[int], vec2: list[int]) -> float:
    if len(vec1) != len(vec2):
        raise ValueError("Векторы должны быть одинаковой длины")
    if not vec1:
        raise ValueError("Векторы не должны быть пустыми")
    return math.sqrt(sum((a - b) ** 2 for a, b in zip(vec1, vec2)) / len(vec1))


def test_mse():
    assert abs(mse([1, 2, 3], [1, 2, 3]) - 0) < 1e-9
    assert abs(mse([1, 2, 3], [2, 3, 4]) - 1) < 1e-9
    assert abs(mse([0, 0, 0], [3, 4, 0]) - math.sqrt((9 + 16 + 0) / 3)) < 1e-9
    assert abs(mse([-1, -2], [1, 2]) - math.sqrt(((2**2) + (4**2)) / 2)) < 1e-9


if __name__ == "__main__":
    test_mse()
```

# Задача 6
Написать функцию, принимающая целое положительное число. Функция должна вернуть
строку вида “(n1**p1)(n2**p2)…(nk**pk)” представляющая разложение числа на простые
множители (если pi == 1, то выводить только ni).
Например, для числа 86240 функция должна вернуть “(2**5)(5)(7**2)(11)”. Написать тесты для кода

```python
def prime_factors(n: int) -> str:
    if n <= 0:
        raise ValueError("Число должно быть положительным")

    factors = {}
    divisor = 2
    while n > 1:
        count = 0
        while n % divisor == 0:
            n //= divisor
            count += 1
        if count > 0:
            factors[divisor] = count
        divisor += 1
        if divisor * divisor > n and n > 1:
            factors[n] = 1
            break

    result = ""
    for prime, power in sorted(factors.items()):
        if power == 1:
            result += f"({prime})"
        else:
            result += f"({prime}**{power})"
    return result


def test_prime_factors():
    assert prime_factors(86240) == "(2**5)(5)(7**2)(11)"
    assert prime_factors(360) == "(2**3)(3**2)(5)"
    assert prime_factors(13) == "(13)"
    assert prime_factors(100) == "(2**2)(5**2)"
    assert prime_factors(1) == ""


if __name__ == "__main__":
    test_prime_factors()
```

# Задача 7
Написать функцию, принимающая целое число n, задающее количество кубиков. Функция
должна определить, можно ли из данного кол-ва кубиков построить пирамиду, то есть
можно ли представить число n как 1^2+2^2+3^2+…+k^2. Если можно, то функция должна
вернуть k, иначе строку “It is impossible”. Написать тесты для кода

```python
def pyramid(n: int):
    if n <= 0:
        raise ValueError("Число должно быть положительным")

    total = 0
    k = 0
    while total < n:
        k += 1
        total += k ** 2
        if total == n:
            return k
    return "It is impossible"


def test_pyramid_height():
    assert pyramid(1) == 1
    assert pyramid(5) == 2
    assert pyramid(14) == 3
    assert pyramid(30) == 4
    assert pyramid(55) == 5
    assert pyramid(15) == "It is impossible"
    assert pyramid(2) == "It is impossible"


if __name__ == "__main__":
    test_pyramid_height()
```

# Задача 8
Функция принимает на вход положительное число и определяет является ли оно сбалансированным, т.е. сумма цифр до средних равна сумме цифр после. Средними в случае нечетного числа цифр считать одну цифру, в случае четного - две средних. Написать тесты для кода

```python
def is_balanced_number(n: int) -> bool:
    if n <= 0:
        raise ValueError("Число должно быть положительным")

    digits = [int(d) for d in str(n)]
    length = len(digits)

    if length == 1:
        return True

    mid = length // 2
    if length % 2 == 0:
        left = digits[:mid - 1]
        right = digits[mid + 1:]
    else:
        left = digits[:mid]
        right = digits[mid + 1:]

    return sum(left) == sum(right)


def test_is_balanced_number():
    assert is_balanced_number(1234006) is True
    assert is_balanced_number(12321) is True
    assert is_balanced_number(123321) is True
    assert is_balanced_number(123456) is False
    assert is_balanced_number(7) is True
    assert is_balanced_number(1230321) is True


if __name__ == "__main__":
    test_is_balanced_number()
```
